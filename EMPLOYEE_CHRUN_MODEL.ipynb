```python
import pandas as pd
import numpy as np
```

CREATING DATAFRAME

---



```python
data=pd.read_csv('/content/HR_comma_sep (1).csv')
data.head(115)
```

CONVERTING QUALTITATIVE ATTRIBUTES TO QUANTITATIVE ATTRIBUTES


---



```python
from sklearn import preprocessing
label=preprocessing.LabelEncoder()
data['salary']=label.fit_transform(data['salary'])
data['Department']=label.fit_transform(data['Department'])
```

```python
data.head(115)
```

```python
  y=data['left']
  X=data.drop('left',axis=1)
```

```python
X.head()
```

SPLITTING THE DATASET INTO TRAINING SET AND TESTING SET

---



```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=10)
```

DEPLOY MODEL

---



```python
from sklearn.neural_network import MLPClassifier
cls=MLPClassifier(hidden_layer_sizes=(100,5),verbose=True,random_state=10,learning_rate_init=0.01)
cls.fit(X_train,y_train)
y_pred=cls.predict(X_test)
```

EVALUATION METRICS

---



```python
from sklearn.metrics import accuracy_score
Accuracy=accuracy_score(y_test,y_pred)
print(Accuracy)
```

```python
error=(1-Accuracy)
print(error)
```
