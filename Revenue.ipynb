```python
data['Date'] = pd.to_datetime(data['Date'], format = "%d/%m/%Y")
```

```python
dat1=data[['Line Id','Day','Date','Name','Total Revenue (F) = A + C + D + E + E1 + E2']].copy()
dat1['year']=dat1['Date'].dt.year
dat1['month']=dat1['Date'].dt.month
dat1['dayno']=dat1['Date'].dt.day
```

```python
dat2=dat1.drop(['Date'],axis=1)
dat2
```

```python
dat2.rename(columns = {'Total Revenue (F) = A + C + D + E + E1 + E2':'Revenue'}, inplace = True)
```

```python
print(dat2[dat2['Revenue']==dat2['Revenue'].min()])
```

```python
d = dat2.groupby('Name')
```

```python
sd=d.get_group('01ED-Wimco Nagar Depot Metro')
sd
```

```python
dummy=pd.get_dummies(dat2.Name)
dummy
dam=pd.get_dummies(dat2.Day)
dam
```

```python
dat3=pd.concat([dat2,dummy,dam],axis=1)
dat3
```

```python
dat4=dat3.drop(['Day','Name'],axis=1)
```

```python
dat4
```

```python
y=dat4.Revenue
X=dat4.drop('Revenue',axis=1)
```

```python
from sklearn.model_selection import train_test_split,ShuffleSplit,cross_val_score
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
!pip install lightgbm
```

```python
from lightgbm import LGBMRegressor
lr=LGBMRegressor()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=10)
cross_val_score(LGBMRegressor(),X,y,cv=cv)
```

```python
line1_col=['0101-Washermanpet',
       '0103-Mannadi', '0105-High Court', '0109-Government Estate', '0111-LIC',
       '0113-Thousand Lights', '0115-AG-DMS', '0117-Teynampet',
       '0119-Nandanam', '0121-Saidapet', '0123-Little Mount', '0125-Guindy',
       '0129-Nanganallur Road', '0131-Meenambakkam',
       '0133-Chennai International Airport', '01ED-Wimco Nagar Depot Metro',
       '01EF-Wimco Nagar Metro', '01F1-Thiruvotriyur Metro',
       '01F3-Thiruvotriyur Theradi Metro', '01F5-Kaladipet Metro',
       '01F7-Tollgate Metro', '01F9-New Washermenpet Metro',
       '01FB-Tondiarpet Metro', '01FD-Thiagaraya College Metro']
line2_col=['0201-Puratchi Thalaivar Dr. M.G. Ramachandran Central', '0203-Egmore',
       '0205-Nehru Park', '0207-Kilpauk', '0209-Pachaiyappas College',
       '0211-Shenoy Nagar', '0213-Anna Nagar East', '0215-Anna Nagar Tower',
       '0217-Thirumangalam', '0219-Koyambedu',
       '0221-Puratchi Thalaivi Dr. J. Jayalalithaa CMBT', '0223-Arumbakkam',
       '0225-Vadapalani', '0227-Ashok Nagar', '0229-Ekkattuthangal',
       '0231-Arignar Anna Alandur','0233-St. Thomas Mount']
print(len(line1_col),len(line2_col))
```

```python
def predict_res(Line,year,month,date,station,day):
  sta_id=0
  if Line==1 and station in line1_col:
    sta_id=np.where(X.columns==station)[0][0]
  elif Line==2 and station in line2_col:
    sta_id=np.where(X.columns==station)[0][0]
  else:
    return "Invalid Line_Id or Station_Name"
  day_id=np.where(X.columns==day)[0][0]
  var=np.zeros(len(X.columns))
  var[0]=Line
  var[1]=year
  var[2]=month
  var[3]=date
  if sta_id>=0:var[sta_id]=1
  if day_id>=0:var[day_id]=1
  return lr.predict([var])[0]
```

```python
print(predict_res(1,2023,8,4,'0131-Meenambakkam','Friday'))
```
