```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix,plot_confusion_matrix
import numpy as np
import pandas 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
def pred(X_train,X_test,y_train,y_test,max_depth,criterion="gini",min_samples_split:int = 2,min_samples_leaf=1):
  TEE=tree.DecisionTreeClassifier(max_depth=max_depth,\
                                  criterion=criterion,\
                                  min_samples_split=min_samples_split,\
                                  min_samples_leaf=min_samples_leaf,\
                                  splitter="best",\
                                  random_state=0,\
                                  )
  
  TEE=TEE.fit(X_train,y_train)
  Y_prd=TEE.predict(y_test)
  print(accuracy_score(Y_prd,y_test))




def give_ex(n,x_train,x_test,y_train,y_test):
  y_train=y_train.reshape(-1,1)
  y_test=y_test.reshape(-1,1)
  pred(x_train,x_test,y_train,y_test,max_depth=n)
  
```

```python
iris=load_iris()
X,Y=iris.data,iris.target
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)
iris_data=pandas.DataFrame(\
                data=np.hstack((X,Y.reshape(Y.shape[0],1),[[iris["target_names"][int(classIdx)]] for classIdx in Y])),\
                columns=["sepal_length","sepal_width","petal_length","petal_width","class","className"])
                       
iris_data.sample(n=10,random_state=1)
```
